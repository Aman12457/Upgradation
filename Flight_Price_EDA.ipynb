```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_excel('flight_price.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Airline'].value_counts()
```

```python
## Feature Enginering
## let's start with the date of journey
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
## drop 'date of journey' col
df.drop('Date_of_Journey',axis=1 , inplace=True)
```

```python
df.head()
```

```python
df['arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
```

```python
df['arrival_min'] = df['Arrival_Time'].str.split(':').str[1].str.split(' ').str[0]
```

```python
df['arrival_hour'] = df['arrival_hour'].astype(int)
df['arrival_min'] = df['arrival_min'].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df.drop('Arrival_Time' , axis=1 , inplace=True)
```

```python
df.head()
```

```python
## converting departure time into dept_hour and dept_min
df['dept_hour'] = df['Dep_Time'].str.split(':').str[0]
df['dept_min'] = df['Dep_Time'].str.split(':').str[1]
```

```python
df['dept_hour'] = df['dept_hour'].astype(int)
df['dept_min'] = df['dept_min'].astype(int)
```

```python
df.drop('Dep_Time' , axis=1 , inplace=True)
```

```python
df.head(2)
```

```python
## Total_Stops turn
df['Total_Stops'].unique()
```

```python
df['Total_Stops'].isnull().sum()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1})
```

```python
df.info()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df['Total_Stops'].value_counts()
```

```python
df.head(2)
```

```python
## no need for route here
df.drop('Route' , axis=1 , inplace=True)
```

```python
df.head(1)
```

```python
## conversion of duration
df['duration_hour'] = pd.to_numeric(df['Duration'].str.split(' ').str[0].str.split('h').str[0] , errors= 'coerce')
df['duration_min'] = pd.to_numeric(df['Duration'].str.split(' ').str[1].str.split('m').str[0] , errors = 'coerce')
```

```python
df.info()
```

```python
df['duration_min'].unique()
```

```python
df['Duration'].isnull().sum()
```

```python
df['duration_hour'].fillna(0 , inplace=True)
df['duration_min'].fillna(0 , inplace=True)
```

```python
df['duration_hour'].isnull().sum()
```

```python
df.drop('Duration' , axis=1 , inplace= True)
```

```python
df.head(1)
```

```python
## converting additional info into numerical feature vector
df['Additional_Info'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
## feed the above values to the encoder
ohe.fit_transform(df[['Source','Destination','Airline']]).toarray()
```

```python
encoded_df = pd.DataFrame(ohe.fit_transform(df[['Source','Destination','Airline']]).toarray() , columns=ohe.get_feature_names_out())
```

```python
final_df = pd.concat([encoded_df , df] , axis =1)
```

```python
final_df.head(2)
```

```python
final_df.drop(['Source','Destination','Airline'] , axis=1 , inplace=True)
```

```python
final_df.info()
```

```python
final_df.drop('Additional_Info', axis=1 , inplace=True)
```
