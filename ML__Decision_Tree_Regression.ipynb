```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[: ,1:-1].values
Y = dataset.iloc[: , -1].values
```

```python
X
```

```python
# Training the Decision Tree model on whole dataset
from sklearn.tree import DecisionTreeRegressor
regressor  = DecisionTreeRegressor(random_state = 0)
regressor.fit(X , Y)
```

```python
# Predicting the result from the above model.
regressor.predict([[6.5]])
```

```python
#Visualizing the Decision Tree Model results
plt.scatter(X, Y , color='red')
plt.plot(X , regressor.predict(X), color='blue')
plt.title('Truth or Bluf(Decision Tree Model)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
#Visualizing Decision Tree Model for higher resolution and smoother curve
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid),1)
plt.scatter(X , Y, color='red')
plt.plot(X_grid , regressor.predict(X_grid), color='blue')
plt.title('Truth or Bluf(Decision Tree Model)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
